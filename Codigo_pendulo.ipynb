```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.integrate import odeint

#par√¢metros
g = 9.8
m = 0.0058
l = 2
t = np.arange(0,10,0.01)

def eqdif(y,t):
    ùúÉ = y[0]
    ùúî = y[1]
    dùúÉdt = ùúî
    dùúîdt = (-g*math.sin(ùúÉ))/l
    return[dùúÉdt,dùúîdt]

iniciais10 = [math.radians(10), 0]
sol10 = odeint(eqdif, iniciais10, t)

# convers√£o

eixoX = []
eixoY = []

for ùúÉ in sol10[:,0]:
    eixoX.append(-l*math.cos(ùúÉ))
    
for ùúÉ in sol10[:,0]:
    eixoY.append(l*math.sin(ùúÉ))
plt.title('Trajet√≥ria do p√™ndulo com 10¬∞ iniciais ')    
plt.grid(True)
plt.xlabel('metros')
plt.ylabel('metros')
plt.plot(eixoY, eixoX)
plt.show()
```
