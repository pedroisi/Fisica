```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.integrate import odeint

#parâmetros
g = 9.8
m = 0.0058
l = 2
t = np.arange(0,10,0.01)

def eqdif(y,t):
    𝜃 = y[0]
    𝜔 = y[1]
    d𝜃dt = 𝜔
    d𝜔dt = (-g*math.sin(𝜃))/l
    return[d𝜃dt,d𝜔dt]

iniciais10 = [math.radians(10), 0]
sol10 = odeint(eqdif, iniciais10, t)

# conversão

eixoX = []
eixoY = []

for 𝜃 in sol10[:,0]:
    eixoX.append(-l*math.cos(𝜃))
    
for 𝜃 in sol10[:,0]:
    eixoY.append(l*math.sin(𝜃))
plt.title('Trajetória do pêndulo com 10° iniciais ')    
plt.grid(True)
plt.xlabel('metros')
plt.ylabel('metros')
plt.plot(eixoY, eixoX)
plt.show()
```

```python
iniciais30 = [math.radians(30), 0]
sol30 = odeint(eqdif, iniciais30, t)

# conversão

eixoX = []
eixoY = []

for 𝜃 in sol30[:,0]:
    eixoX.append(-l*math.cos(𝜃))
    
for 𝜃 in sol30[:,0]:
    eixoY.append(l*math.sin(𝜃))

plt.title('Trajetória do pêndulo com 30° iniciais ')    
plt.grid(True)
plt.xlabel('metros')
plt.ylabel('metros')
plt.plot(eixoY, eixoX)
plt.show()
```
